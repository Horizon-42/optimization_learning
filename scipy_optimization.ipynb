### Nelder-Mead Simplex algorithm
Simplex, means triangle or tetrahedron in any dimensions;
1-D, for line segment, 2-D, for triangle, 3-D, tetrahedran, n-D, n+1 vertices.
This method find a **minmum** without using derivatives;
Moves a triagnle around the surface to look for a low point with moves like:
- Reflcection, flip the worst point across the center.
- Expansion, go farther if that helped.
- Contraction, move less far if reflection was not good.
- Shrink, make the whole simplex smaller around the best point.

```python
import numpy as np 
from scipy.optimize import minimize

def rosen(x):
    """The Rosenbrock function"""
    return sum(100.0*(x[1:]-x[:-1]**2.0)**2.0 + (1-x[:-1])**2.0)

# Example usage:
x0 = np.array([1.3, 0.7, 0.8, 1.9, 1.2])
# Minimize the Rosenbrock function starting from the initial guess x0
#  xatol is the absolute error in xopt between iterations that is acceptable for convergence.
result = minimize(rosen, x0, method='nelder-mead', options={'xatol': 1e-8, 'disp': True})
print("Optimal point:", result.x)
print("Function value at optimal point:", result.fun)
```

```python
# Optimization with additional args
def rosen_with_args(x, a, b):
    """The Rosenbrock function with additional arguments"""
    return sum(a*(x[1:]-x[:-1]**2.0)**2.0 + (1-x[:-1])**2.0) + b

# Example usage with additional arguments:
x0 = np.array([1.3, 0.7, 0.8, 1.9, 1.2])
# using args parameter
result = minimize(rosen_with_args, x0, args=(100.0, 1.0), method='nelder-mead', options={'xatol': 1e-8, 'disp': True})
print("Optimal point with args:", result.x)

# uding wrap funtion
def rosen_with_args_wrap(x):
    return rosen_with_args(x, 100.0, 1.0)
result_wrap = minimize(rosen_with_args_wrap, x0, method='nelder-mead', options={'xatol': 1e-8, 'disp': True})
print("Optimal point with wrap function:", result_wrap.x)

# using partial function
from functools import partial
rosen_with_args_partial = partial(rosen_with_args, a=100.0, b=1.0)
result_partial = minimize(rosen_with_args_partial, x0, method='nelder-mead', options={'xatol': 1e-8, 'disp': True})
print("Optimal point with partial function:", result_partial.x)

# using lambda function
rosen_with_args_lambda = lambda x: rosen_with_args(x, 100.0, 1.0)
result_lambda = minimize(rosen_with_args_lambda, x0, method='nelder-mead', options={'xatol': 1e-8, 'disp': True})
print("Optimal point with lambda function:", result_lambda.x)
```

### Broyden-Fletcher-Goldfarb-Shanno algorithm


```python
def rosen_der(x):
    """The gradient of the Rosenbrock function"""
    xm = x[1:-1] # x[1] to x[n-2], x_j
    xm_m1 = x[:-2] # x[0] to x[n-3], x_{j-1}
    xm_p1 = x[2:] # x[2] to x[n-1], x_{j+1}
    der = np.zeros_like(x)
    der[1:-1] = 200*(xm-xm_m1**2) - 400*(xm_p1-xm**2)*xm - 2*(1-xm)
    der[0] = -400*x[0]*(x[1]-x[0]**2) - 2*(1-x[0])
    der[-1] = 200*(x[-1]-x[-2]**2)
    return der

# Minimize the Rosenbrock function using the BFGS algorithm, which uses the gradient
result_bfgs = minimize(rosen, x0, method='BFGS', jac=rosen_der, options={'disp': True})
print("Optimal point with BFGS:", result_bfgs.x)
print("Function value at optimal point with BFGS:", result_bfgs.fun)

# could use jac = True to reduce redundant function evaluations
```
